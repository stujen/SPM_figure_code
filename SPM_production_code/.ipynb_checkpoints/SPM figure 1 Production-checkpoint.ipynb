# SPM figure 1

Code to produce standard version of SPM figure 1, SR1.5. Written by Stuart Jenkins (stuart.jenkins@wadham.ox.ac.uk) and Richard Millar (richard.millar@ouce.ox.ac.uk).

Switches at top allow user to generate each plume individually or all three together, and to change between IPCC AR5 baseline (1850-1900) and nearterm reference period (2006-2015).

__Order of run__:

1) Import statements (external packages followed by model import and colour definitions)

2) Import and rebaseline historical temperature observations, calculate AWI best estimate using Piers' updated radiative forcing and Karsten Haustein real-time warming index method (K. Haustein et al., 2017)

3) Define parameters for climate, and scenario generation. Use these to define scenarios. Methodology for scenarios follows Millar paper closely (R. Millar et al., 2017)

4) Run through the model to find likely range of temperature responses

5) Plot in style of SPM figure 1

### Switches:

```python
# What plume do you want to plot? Options - 'grey', 'blue', 'purple' or 'all'
plume = 'all'

# What baseline period do you want to use? Options - 'ipcc_ar5_preind' (1850-1900) 
#      or 'nearterm' (2006-2015)
baseline_choice = 'ipcc_ar5_preind'
```

## Code for figure production

### 1. Import statements

First we import the third party packages required for this script. This is followed by importing code from the subfolder 'Support_codes'. These codes contain functions to generate the rest of the figure. 

_fair_\__scm.py_ : FaIR simple climate model code. The function __fair_scm__ runs forward from emissions, concentrations or radiative forcing to temperatures. __fair_scm_emsback__ runs backwards, from temperatures/concentrations to emissions. 

_temperature_\__import_\__and_\__baseline.py_ : Code for importing the temperature datasets we use over the historical period. Currently the 4 historical temperature observation datasets we use are HadCRUT4, HadCRUT4-Cowtan-Way, NOAA and GISTEMP. The JMA and Berkley datasets are also consistent with the mean of the datasets. They are imported and rebaselined to a common reference period with __temp_import__, and a min/max/mean of the datasets is calcualted with __calc_mean_min_max__. __calc_gwi__ is used to calculate the most likely contributions from anthropogenic and natural sources using Ordinary Least Squares (OLS) regression. 

_forcing_\__imports.py_ : Code for importing the historical radiative forcing timeseries used in the subpanels and to calculate the plume temperature responses in the main panel. __load_rf_comps__ and __load_contrails_rf__ imports the RF timeseries. The RF timeseries are then scaled to match the AR5 2011 most-likely RF values. The guassian radiative forcers are sampled to span their AR5 likely ranges (__create_gauss_scalingfactors__). Then the aerosol forcers are scaled to match the 2017 FaIR derived anthropogenic warming likely range, consistent with the AR5 derived range and the observations (__create_aerosol_scalingfactors__). The CO2 emissions timeseries are inverted from the radiative forcing timeseries (imported with __load_rf_comps__). Carbon cycle parameters are tuned so they span the ratios in CMIP5 and diagnose the 2017 GCP CO2 emissions value (__calc_cc_uncert__). 

_SPM_\__scenario_\__codes.py_ : Code to form the scenarios and generate the plumes. Carbon emissions are diagnosed by inverting CO2 radiative forcing timeseries from RCP8.5, and imposing a linear decline in emissions to zero by 2055 (or 2040). Non-CO2 RF scaled from RCP8.5 in the past and have a defined functional form attached after 2020 (increasing till 2030 and declining thereafter) (__linear_decline_tozero__).

```python
# External packages required for figure production
# -------------------------------------------------
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Patch
from matplotlib.legend_handler import HandlerTuple
# -------------------------------------------------

# Internal imports - support code for main figure production
# -------------------------------------------------
# Import the FaIR climate-carbon-cycle model from python file
from Support_codes.fair_scm import *
# Import and prepare the historical temperature observations data
from Support_codes.temperature_import_and_baseline import *
# Calculate appropriate parameter sets for scenarios
from Support_codes.forcing_imports import *
# Calculate appropriate parameter sets for scenarios
from Support_codes.SPM_scenario_creator import *
# -------------------------------------------------

# Load colour palette for vermillion colour definition (used for historical anthropogenic temperature range)
#    All other colours are standard predefined colour within matplotlib.
def load_colpal():
    rgb_list = np.array([[213,94,0]]) / 255.
    rgb_names = ['vermillion']
    rgbs = {rgb_names[x]:tuple(rgb_list[x]) for x in range(0,len(rgb_list))}
    return rgbs
cols = load_colpal()
```

### 2. Import and re-baseline historcial temperatures, calcualte AWI


Code in this section uses the 'temperature_import_and_baseline.py' support code.

Import the 4 temperature datasets (HadCRUT4, HadCRUT4-Cowtan-Way, NOAA and GISTEMP) and rebaseline them to a common reference period (1850-1900). Calculate the mean, max and min of the datasets and fit using OLS regression to model outputs of anthro and natural contributions to temperature observations. All are saved into the dictionary gmst, with keywords of the gmst_names, temp-mean, temp-max, awi, nwi etc.

```python
# Define gmst dataset names
gmst_names = ['HadCRUT4','GISTEMP','NOAA','Cowtan-Way']
# define the baseline year range, and common reference range
base_low=1850.
base_high=1900.
# define variable representing frequency of temp. observations ('mon' = monthly)
temp_freq='mon'
# ----------------------------------------------------



# Import the datasets (HadCRUT4, HadCRUT4-CW, NOAA and GISTEMP) into dictionary 'gmst'
gmst = temp_import()
# Find min, max and mean of temperature observation datasets
gmst = calc_mean_min_max(gmst)
# calculate the AWI and NWI for each GMST dataset relative to the baseperiod and also for the mean of the datasets 
for key in gmst_names:
    awi, nwi = calc_gwi(gmst[key]['Temp'],gmst[key]['Years'],reg_type=temp_freq,
                        base_low=base_low,base_high=base_high, name=key)
    # express AWI and NWI relative the mid-point of the period
    gmst[key]['AWI'] = awi 
    gmst[key]['NWI'] = nwi 

# calculte for the mean of the datasets
awi, nwi = calc_gwi(gmst['Temp-mean'],gmst['HadCRUT4']['Years'],reg_type=temp_freq,
                    base_low=base_low,base_high=base_high, name='Mean temp.')
gmst['Temp-mean-AWI'] = awi 
gmst['Temp-mean-NWI'] = nwi
```

### 3. Define parameters for scenario creation

This is the main section for design of the scenarios. The scenarios are scaled and tuned to match the historical observations record, and to give a range of possible future pathways using the likely range of climate parameters.

Climate parameters TCR and ECS are sampled to span the AR5 likely range in their distributions (cdf values are 17%, 33%, 50%, 67%, 83%).

Scaling is performed first to match the AR5 reported most likely foricng contributions, then to span the gaussian foricng likely range, and finally to aerosol forcing to obtain a consistent 2017 likely warming range. 

Carbon cycle parameters are tuned to diagnose the 2017 emissions consistent with GCP emissions, with the relevent climate parameters. 

Finally, TCREs are calculated for each set of parameters and all parameters are printed to the screen.

```python
# Set directory where datafiles are stored
datadir='./Data/'
# Set shape of TCR CDF
shape='gaussian'
# set key parameters
a=3.74/np.log(2.0)
c0=278.
# set baseline period for FaIR model runs
base_low=1861.
base_high=1880.
# Set FaIR model temperature boxes parameters
d=np.array([409.5,8.4])

years = np.arange(1765.0,2501.0)


rf_scens = {}
#Load the standard RCP8.5 scenario
scen_names = ['RCP8.5']
for name in scen_names:
    
    # load forcing from file
    rf_file = datadir+'RCP85_MIDYEAR_RADFORCING.csv'
    # load the forcing components from the file
    rf_comps, rf_nat, ar5_comp_scale = load_rf_comps(rf_file,ar5_scale=True)
    # find total natural forcing by adding components
    rf_nat = np.sum(rf_nat,axis=0) 
    
    # combing non-CO2 gaussian forcing components to find total gaussian non-CO2 RF 
    rf_scens = comb_rf_comps(rf_scens,rf_comps,name)

    #Load the historical contrail rf
    rf_cont = load_contrails_rf(years)
    
# Sample TCR and ECS distributions to find values spanning central tercile and 2/3rds likely range
cdfs, TCR, ECS = create_climresp_dist()
# find the gaussian non-CO2 forcing scaling factors
sf_gauss = create_gauss_scalingfactors(cdfs)
# find the aerosol forcing scaling factors (so they contribute the required fraction of total warming in 2017)  
sf_aero, rf_aero_2011, rf_tot_2011 = create_aerosol_scalingfactors(cdfs, TCR, ECS, rf_scens, sf_gauss, rf_cont, rf_nat)
# find the carbon cycle parameters to diagnose the GCP annual CO2 emissions quantity in 2017 (11.7GtC)
s_temp, s_cu, iirf_100_preind = calc_cc_uncert(TCR, ECS, cdfs, rf_scens, sf_aero, sf_gauss, rf_cont, rf_nat)
# Calculate the TCRE for each TCR and ECS set
TCRE = calc_tcre(TCR, ECS, iirf_100_preind, s_cu, s_temp)
# Print the key parameters for review
print_parameters(cdfs, TCR, ECS, TCRE, rf_aero_2011, rf_tot_2011, sf_gauss, sf_aero, iirf_100_preind, s_cu, s_temp)
```

### 4. Define scenario

Here, we run the function __linear_decline_tozero__ to define the scenarios and resulting temperature response plumes. Temperature responses are calculated with the FaIR simple climate model with the likely range of parameters as calculated above for each scenario. 

Different scenarios are:

- grey: linear decline in CO2 emissions from 2020 level to zero by 2055, with rising non-CO2 forcing until 2030 (something which is seen in all ambitious mitigation scenarios, and is a cosequence of rapid reductions in aerosol forcing and continuing OWMGHG forcing), and declining thereafter. 

- blue: faster reductions in CO2 emissions, declining from 2020 level to zero by 2040. 

- purple: CO2 emissions reduce to zero by 2055 and non-CO2 foricng increases till 2030, where it then stays constant thereafter. 

```python
# predefine some standard solution so something will always plot...
rf_targs = [0.4]
co2_times_te = np.array([2020,2055,2075])
co2_fracs_te = np.array([1.0,0.0,0.0])
te = {}

if plume == 'grey':
    # Lower non-CO2 forcing trajectory
    rf_targs = [0.4]
    # zero CO2 emissions by 2055
    co2_times_te = np.array([2020,2055,2075])
    co2_fracs_te = np.array([1.0,0.0,0.0])
elif plume == 'blue':
    # Lower non-CO2 forcing trajectory
    rf_targs = [0.4]
    # zero CO2 emissions by 2040
    co2_times_te = np.array([2020,2040,2055,2075])
    co2_fracs_te = np.array([1.0,0.0,0.0,0.0])
elif plume == 'purple':
    # Higher non-CO2 forcing trajectory
    rf_targs = [0.8]
    # zero CO2 emissions by 2055
    co2_times_te = np.array([2020,2055,2075])
    co2_fracs_te = np.array([1.0,0.0,0.0])
elif plume == 'all':
    # Lower non-CO2 forcing trajectory
    rf_targs = [0.4, 0.8]
    # zero CO2 emissions by 2055
    co2_times_te = np.array([2020,2055,2075])
    co2_fracs_te = np.array([1.0,0.0,0.0])
    
    # Lower non-CO2 forcing trajectory
    rf_targs_te2 = [0.4]
    # zero CO2 emissions by 2040
    co2_times_te2 = np.array([2020,2040,2055,2075])
    co2_fracs_te2 = np.array([1.0,0.0,0.0,0.0])

# Find the pathway and save to dictionary 'te'
temp_zero_te, llems_zero_te, slcp_zero_te = linear_decline_tozero(TCR, ECS, sf_gauss, sf_aero, rf_scens, rf_cont, 
                                                                       rf_nat, iirf_100_preind, s_temp, s_cu, 
                                                                       slcp_mode='parab',rf_targs=rf_targs,
                                                                       co2_times=co2_times_te,
                                                                       co2_fracs=co2_fracs_te, 
                                                                       tcrecs_index=[0,1,2,3,4])

te = {'years':years,'temp_zero':temp_zero_te,'llems_zero':llems_zero_te,'slcp_zero':slcp_zero_te}

te2 = copy.deepcopy(te)
if plume == 'all':
    te2['temp_zero'],te2['llems_zero'], te2['slcp_zero'] = linear_decline_tozero(TCR, ECS, sf_gauss, sf_aero, rf_scens, rf_cont, 
                                                                           rf_nat, iirf_100_preind, s_temp, s_cu, 
                                                                           slcp_mode='parab',rf_targs=rf_targs_te2,
                                                                           co2_times=co2_times_te2,
                                                                           co2_fracs=co2_fracs_te2, 
                                                                           tcrecs_index=[0,1,2,3,4],
                                                                           temp_zero=te2['temp_zero'],
                                                                           llems_zero=te2['llems_zero'],
                                                                           slcp_zero=te2['slcp_zero'])

    #te2 = {'years_te2':years,'temp_zero_te2':temp_zero_te2,'llems_zero_te2':llems_zero_te2,'slcp_zero_te2':slcp_zero_te2}
###################################################
###################################################

# import GCP carbon emissions data
gcp_data = np.genfromtxt('./Data/gcp_data_2017_prelim.csv',delimiter=',',
                    dtype=np.dtype({'names':['Year','FFI','LUC'],'formats':3*[np.float]}))
# total carbon emissions are fossil carbon + LUC
gcp_ems = gcp_data['FFI'] + gcp_data['LUC']

########################################################
### calculate cumulative carbon emissions trajectory ###
########################################################
# Find the cumulative CO2 emissions pathway by summing the annual emissions trajectory (change units to TtCO2)
if plume != 'all':
    cum_high = 44./12000*(np.cumsum(te['llems_zero']['RCP8.5'][0][1]))
    # rebaseline to be zero in 1870
    cum_high = cum_high[te['years'][:]>=1950] - cum_high[te['years'][:]==1870]
else:
    cum_high_grey = 44./12000*(np.cumsum(te['llems_zero']['RCP8.5'][0][1]))
    # rebaseline to be zero in 1870
    cum_high_grey = cum_high_grey[te['years'][:]>=1950] - cum_high_grey[te['years'][:]==1870]
    
    cum_high_blue = 44./12000*(np.cumsum(te2['llems_zero']['RCP8.5'][0][1]))
    # rebaseline to be zero in 1870
    cum_high_blue = cum_high_blue[te2['years'][:]>=1950] - cum_high_blue[te2['years'][:]==1870]

# get subset of data up to present day (to plot as darker line on subpanel)
x1 = te['years'][:]<=2017
x2 = te['years'][:]>=1950
cum_high_topresentday = 44./12000*(np.cumsum(te['llems_zero']['RCP8.5'][0][1]))
cum_high_topresentday = cum_high_topresentday[x1*x2] - cum_high_topresentday[te['years'][:]==1870]

# find cumulative GCP emissions and baseline to 1870
gcp_ems[0] = 0
cum_gcp = 44./12000*np.cumsum(gcp_ems)
cum_gcp = cum_gcp[gcp_data['Year']>=1850] - cum_gcp[gcp_data['Year']==1870]

cum_gcp_years = gcp_data['Year'][gcp_data['Year']>=1850]
########################################################
########################################################

##################################################
### calculate the non-CO2 forcing contribution ###
##################################################

# it is already done in production of scenario

##################################################
```

### 5. Plot figure

A lot of this main script is devoted to defining an appropriate figure shape and populating it with the information seen in SPM figure 1. Below is a large section of code which formats the figure and plots the final figure, depending on the switch positions at the top of the page.

Main structure is:

1) define the figure structure (in top hashed section) and add on historical temperature record.

2) set the lines in the left subpanel (annual CO2 emissions). Historical are plotted GCP emissions timeseries, future are inverted emissions from RCP8.5 RF timeseries (harmonized to present day GCP emissions with appropriate carbon cycle parameter choice).

3) set the lines in the middle subpanel (cumulative CO2 emissions). Cumulative emissions are the running sum of all annual emissions up to that year. 

4) set the lines in the right subpanel (non-CO2 forcing contribution). Non-CO2 forcing has contributions from OWMGHGs, ozone, aerosols, black carbon, vapour (H20) and Landuse change.

5) draw the temperature plumes on which correspond to each scenarios likely response to the inputs.

```python
###############################################################################
###############################################################################
# DEFINE THE SPM FIGURE 1 SHAPE
###############################################################################
###############################################################################
                                                                                                    
# define a figure shape - locations and sizes of axes
main_xstart = 0.13; main_xwidth = 0.79; main_ystart = 0.50; main_ywidth = 0.42
left_xstart = 0.13; left_xwidth = 0.23; left_ystart = 0.11; left_ywidth = 0.30
mid_xstart = 0.41; mid_xwidth = 0.23; mid_ystart = 0.11; mid_ywidth = 0.30
right_xstart = 0.69; right_xwidth = 0.23; right_ystart = 0.11; right_ywidth = 0.30

# define the figure
fig = plt.figure(figsize=(12,8))
# define the left subplot
ax_left = fig.add_axes([left_xstart,left_ystart,left_xwidth,left_ywidth])
ax_left.spines['right'].set_visible(False)
ax_left.spines['top'].set_visible(False)
ax_left.tick_params(colors='white')
ax_left.set_xlim(1960,2100)
ax_left.set_xticks([1980,2020,2060,2100],minor = False)
ax_left.set_xticks([1960,2000,2040,2080],minor = True)
ax_left.set_ylim(0,60)
ax_left.set_yticks([0,20,40,60],minor = True)
ax_left.set_yticks([0,10,20,30,40,50,60],minor = False)
left_text = ax_left.text(0.03,1.0,'Annual CO$_{2}$ emissions',fontsize=12,
                         transform=ax_left.transAxes,ha='left',color='white')
# define the middle subplot
ax_middle = fig.add_axes([mid_xstart,mid_ystart,mid_xwidth,mid_ywidth])
ax_middle.spines['right'].set_visible(False)
ax_middle.spines['top'].set_visible(False)
ax_middle.set_xlim(1960,2100)
ax_middle.set_xticks([1980,2020,2060,2100],minor = False)
ax_middle.set_xticks([1960,2000,2040,2080],minor = True)
ax_middle.set_ylim(0,3.5)
ax_middle.set_yticks([0,1,2,3],minor = False)
ax_middle.set_yticks([0.5,1.5,2.5,3.5],minor = True)
ax_middle.tick_params(colors='white')
mid_text = ax_middle.text(0.03,1.0,'Cumulative CO$_{2}$ emissions', 
               fontsize=12,transform=ax_middle.transAxes, 
               ha='left', color='white')
# define the right subplot
ax_right = fig.add_axes([right_xstart,right_ystart,right_xwidth,right_ywidth])
ax_right.spines['right'].set_visible(False)
ax_right.spines['top'].set_visible(False)
ax_right.set_xlim(1960,2100)
ax_right.set_xticks([1980,2020,2060,2100],minor = False)
ax_right.set_xticks([1960,2000,2040,2080],minor = True)
ax_right.set_ylim(0,3.5)
ax_right.set_yticks([0,1,2,3],minor = False)
ax_right.set_yticks([0.5,1.5,2.5,3.5],minor = True)
ax_right.tick_params(colors='white')
right_text = ax_right.text(0.03,1.0,'Non-CO$_{2}$ climate forcing', 
               fontsize=12,transform=ax_right.transAxes, 
               ha='left', color='white')
# define the main subplot
ax_main_fig = fig.add_axes([main_xstart,main_ystart,main_xwidth,main_ywidth])
ax_main_fig.spines['right'].set_visible(False)
ax_main_fig.spines['top'].set_visible(False)
ax_main_fig.set_xlim(1960,2100)
if baseline_choice == 'ipcc_ar5_preind':
    ax_main_fig.set_ylim(0,2.0)
    ax_main_fig.set_yticks([0,0.5,1.0,1.5,2.0],minor=False)
    ax_main_fig.set_yticks([0.25,0.75,1.25,1.75],minor=True)
    ax_main_fig.set_ylabel(r'Global warming ($^\circ$C) relative to 1850-1900')
elif baseline_choice == 'nearterm':
    ax_main_fig.set_ylim(-1.0,1.0)
    ax_main_fig.set_yticks([-1.0,-0.5,0.0,0.5,1.0],minor=False)
    ax_main_fig.set_yticks([-0.75,-0.25,0.25,0.75],minor=True)
    ax_main_fig.set_ylabel(r'Global warming ($^\circ$C) relative to 2006-2015')

# add the lines which make up the historical temperature timeseries data
if baseline_choice == 'ipcc_ar5_preind':
    ax_main_fig.fill_between(gmst['HadCRUT4']['Years'],gmst['Temp-max'],gmst['Temp-min'],
                         color='grey',alpha=0.3)
    ax_main_fig.plot(gmst['HadCRUT4']['Years'],np.array(gmst['Temp-mean']),color='grey',
                 linestyle='-',linewidth=1, alpha = 0.4)
    ax_main_fig.plot(gmst['HadCRUT4']['Years'], gmst['Temp-mean-AWI'], color=cols['vermillion'],
                 linestyle='-', linewidth=2)
    ax_main_fig.fill_between(gmst['HadCRUT4']['Years'],gmst['Temp-mean-AWI']*(1.2),
                     gmst['Temp-mean-AWI']*(0.8),color=cols['vermillion'],
                     alpha=0.4,linewidth=0)
elif baseline_choice == 'nearterm':
    p_max = gmst['Temp-max'] - np.mean(gmst['Temp-max'][np.logical_and(gmst['HadCRUT4']['Years'][:]>2006,gmst['HadCRUT4']['Years'][:]<2016)])
    p_min = gmst['Temp-min'] - np.mean(gmst['Temp-min'][np.logical_and(gmst['HadCRUT4']['Years'][:]>2006,gmst['HadCRUT4']['Years'][:]<2016)])
    p_mean = gmst['Temp-mean-AWI'] - np.mean(gmst['Temp-mean-AWI'][np.logical_and(gmst['HadCRUT4']['Years'][:]>2006,gmst['HadCRUT4']['Years'][:]<2016)])
    
    ax_main_fig.fill_between(gmst['HadCRUT4']['Years'],np.array(p_mean)*(1.2),np.array(p_mean)*(0.8),color=cols['vermillion'],zorder=5,alpha=0.4,linewidth=0)

    ax_main_fig.fill_between(gmst['HadCRUT4']['Years'],p_max,p_min,color='grey',alpha=0.3)
    ax_main_fig.plot(gmst['HadCRUT4']['Years'],np.array(p_mean),color=cols['vermillion'],linestyle='-',linewidth=2)
###############################################################################
###############################################################################
# END OF MAIN FIGURE CREATION
###############################################################################
###############################################################################






# Add a line showing the position of 2017 and current temperature gradient
# ----------------------------------------------- 
if baseline_choice == 'ipcc_ar5_preind':
    warm_end = np.mean(gmst['Temp-mean-AWI'][np.logical_and(np.array(gmst['HadCRUT4']['Years'])>=2017,
                                                            np.array(gmst['HadCRUT4']['Years'])<2018)])
elif baseline_choice == 'nearterm':
    warm_end = np.mean(np.array(p_mean)[np.logical_and(np.array(gmst['HadCRUT4']['Years'])>=2017,
                                                       np.array(gmst['HadCRUT4']['Years'])<2018)])

# add orange dot at 2017 observed temperature value
ax_main_fig.scatter(2017.5,warm_end,s=30,color='orange',zorder=20)

#Add projection for current warming rate 
proj_short = np.arange(2018,2042)
warm_rate = 0.021
ax_main_fig.plot(proj_short+0.5,warm_end+warm_rate*(proj_short+0.5-2017.5),linestyle='--',
                 dashes=(3, 2),color=cols['vermillion'],linewidth=1.5,zorder=10, label='If current warming rate continues')

# add horizontal line at 1.5 degrees and vertical line at 2017 to main axes
if baseline_choice == 'ipcc_ar5_preind':
    ax_main_fig.axhline(y=1.5,linewidth=0.75,linestyle='--',dashes=(3,3),color='grey')
    ax_main_fig.text(2015,0.43,'2017',fontsize=10,color='black')
    ax_main_fig.plot([2017.5,2017.5],[0,0.40],linestyle='--',dashes=(3,3),linewidth=0.75,color='grey')
    ax_main_fig.plot([2017.5,2017.5],[0.52,warm_end],linestyle='--',dashes=(3,3),linewidth=0.75,color='grey')
elif baseline_choice == 'nearterm':
    ax_main_fig.axhline(y=1.5-0.87,linewidth=1.5,linestyle='-',color='grey')
#     ax_main_fig.axhline(y=0-0.87,linewidth=1.5,color='grey',linestyle='-',zorder=0)
#     ax_main_fig.text(x=1960, y=0.7-0.87, s='Human-induced\nwarming', fontsize=13,color=awi_col)
    
# aded legend for current warming rate line
legend6 = ax_main_fig.legend(loc='upper left', framealpha=1, edgecolor='black')
# -----------------------------------------------









# Set axes parameters for left hand subpanel
# -----------------------------------------------
ax_left.tick_params(colors='black'); left_text.set_color('black')  
ax_left.set_ylabel('Billion tCO$_{2}$/yr',fontsize=10)
# -----------------------------------------------

# Add error bar for 2017 annual GCP emissions value
ax_left.errorbar(2017, 44./12*te['llems_zero']['RCP8.5'][0][1][te['years'][:]==2017], 0.08*44./12*te['llems_zero']['RCP8.5'][0][1][te['years'][:]==2017], fmt='none',elinewidth=1,ecolor='gray',capsize=3)

# plot annual CO2 emissions on left subpanel
line_5, = ax_left.plot(gcp_data['Year'][gcp_data['Year']>=1850],44./12*gcp_ems[gcp_data['Year']>=1850],color='grey',linewidth=2)

if (plume == 'grey') or (plume == 'purple'):
    line_1, = ax_left.plot(te['years'][te['years'][:]>=2017],44./12*te['llems_zero']['RCP8.5'][0][1][te['years'][:]>=2017],color='grey',linewidth=2, alpha = 0.45)
if plume == 'blue':
    # add the trajectory of CO2 emissions to the left and middle subpanels
    values_to_plot = (te['years'][:]>=2017) * (te['years'][:]<=2100)
    blue_line = ax_left.plot(te['years'][values_to_plot],44./12*te['llems_zero']['RCP8.5'][0][1][values_to_plot],color='blue',linewidth=2,alpha=0.6,clip_on=False)
    blue_line[0].set_clip_on(False)
if plume == 'all':
    line_1, = ax_left.plot(te['years'][te['years'][:]>=2017],44./12*te['llems_zero']['RCP8.5'][0][1][te['years'][:]>=2017],color='grey',linewidth=2, alpha = 0.45)
    # add the trajectory of CO2 emissions to the left subpanel
    values_to_plot = (te['years'][:]>=2017) * (te['years'][:]<=2100)
    blue_line = ax_left.plot(te2['years'][values_to_plot],44./12*te2['llems_zero']['RCP8.5'][0][1][values_to_plot],color='blue',linewidth=2,alpha=0.6,clip_on=False)
    blue_line[0].set_clip_on(False)





# -----------------------------------------------
# Set axes parameters for middle subpanel
# -----------------------------------------------
ax_middle.tick_params(colors='black'); mid_text.set_color('black')
ax_right.tick_params(colors='white'); right_text.set_color('white')
ax_middle.set_ylabel('Trillion tonnes CO$_2$',fontsize=10); ax_right.set_ylabel('')
# -----------------------------------------------

# Add error bar showing GCP emissions uncertainty in 2017
ax_middle.errorbar(2017, cum_gcp[cum_gcp_years==2017], 0.11*cum_gcp[cum_gcp_years==2017], fmt='none',elinewidth=1,ecolor='gray',capsize=3)
########################################################
########################################################

x1 = te['years'][:]<=2017
x2 = te['years'][:]>=1950

if (plume == 'grey') or (plume == 'purple'):
    # add cumulative emissions to middle panel
    line_6, = ax_middle.plot(te['years'][te['years'][:]>=1950],cum_high,color='grey',
                            alpha=0.45,linewidth=2)

    line_7, = ax_middle.plot(te['years'][x1*x2],cum_high_topresentday,color='grey',linewidth=2)
if plume == 'blue':
    line6, = ax_middle.plot(te['years'][te['years'][:]>=1950],cum_high,color='blue',linewidth=2,alpha=0.6)
    line_7, = ax_middle.plot(te['years'][x1*x2],cum_high_topresentday,color='grey',linewidth=2)
if plume == 'all':
    line6, = ax_middle.plot(te2['years'][te2['years'][:]>=1950],cum_high_grey,color='grey',linewidth=2,alpha=0.45)
    ax_middle.plot(te2['years'][te2['years'][:]>=1950],cum_high_blue,color='blue',linewidth=2,alpha=0.6)
    line_7, = ax_middle.plot(te['years'][x1*x2],cum_high_topresentday,color='grey',linewidth=2)

    
    
    
    
    
    
    
# -----------------------------------------------
# Set axes parameters for right hand subpanel
# -----------------------------------------------
ax_right.tick_params(colors='black'); right_text.set_color('black')
ax_right.set_ylabel('Watts per square metre',fontsize=10)
# -----------------------------------------------

if (plume == 'grey') or (plume == 'blue'):
    line_8, = ax_right.plot(te['years'],te['slcp_zero']['RCP8.5'][0][1],color='grey',
                            linewidth=2,alpha = 0.45)

    line_9, = ax_right.plot(te['years'][np.logical_and(te['years'][:]<=2011,te['years'][:]>=1960)],
                            te['slcp_zero']['RCP8.5'][0][1][np.logical_and(te['years'][:]<=2011,te['years'][:]>=1960)],
                            color='grey',linewidth=2)
if plume == 'purple':
    line_8, = ax_right.plot(te['years'],te['slcp_zero']['RCP8.5'][0][1],color='red',
                            linewidth=2,alpha = 0.55)

    line_9, = ax_right.plot(te['years'][np.logical_and(te['years'][:]<=2011,te['years'][:]>=1960)],
                            te['slcp_zero']['RCP8.5'][0][1][np.logical_and(te['years'][:]<=2011,te['years'][:]>=1960)],
                            color='grey',linewidth=2)
if plume == 'all':
    line8, = ax_right.plot(te['years'],te['slcp_zero']['RCP8.5'][0][1],color='grey',
                            linewidth=2,alpha = 0.45)
    
    ax_right.plot(te['years'],te['slcp_zero']['RCP8.5'][1][1],color='red',
                            linewidth=2,alpha = 0.55)

    line_9, = ax_right.plot(te['years'][np.logical_and(te['years'][:]<=2011,te['years'][:]>=1960)],
                            te['slcp_zero']['RCP8.5'][1][1][np.logical_and(te['years'][:]<=2011,te['years'][:]>=1960)],
                            color='grey',linewidth=2)

# dont clip line below x-axis
line_9.set_clip_on(False)

#Scale AR5 uncertainty down to likely range 
mu_nco2 = 0.5*(1.13 + 3.33) - 1.68
sig_nco2 = (1.13 - 1.68 - mu_nco2)/(erfinv((0.05)*2 - 1)*np.sqrt(2.0))
nco2_range = sig_nco2 * np.sqrt(2.0) * erfinv(2*(np.array([0.17,0.83])) -1.0) + mu_nco2

# set error bar and allow it to extend below x axis 
#----------------------------------------
e = ax_right.errorbar(2011,mu_nco2,mu_nco2-nco2_range[0],fmt='none',elinewidth=1,ecolor='gray',capsize=3)
for b in e[1]:
    b.set_clip_on(False)
for b in e[2]:
    b.set_clip_on(False)
# -----------------------------------------------







if (plume == 'grey') or (plume == 'blue') or (plume == 'purple'):
    # find the relevent ranges for temperature plume
    plot_mean = te['temp_zero']['RCP8.5'][0][1] 
    plot_uu = te['temp_zero']['RCP8.5'][0][2] 
    plot_ll = te['temp_zero']['RCP8.5'][0][0] 
    plot_ul = te['temp_zero']['RCP8.5'][0][4] 
    plot_lu = te['temp_zero']['RCP8.5'][0][3]
    
    if baseline_choice == 'nearterm':
        plot_mean = express_as_anom(plot_mean, te['years'], base_high=2015, base_low=2006)
        plot_uu = express_as_anom(plot_uu, te['years'], base_high=2015, base_low=2006)
        plot_ll = express_as_anom(plot_ll, te['years'], base_high=2015, base_low=2006)
        plot_ul = express_as_anom(plot_ul, te['years'], base_high=2015, base_low=2006)
        plot_lu = express_as_anom(plot_lu, te['years'], base_high=2015, base_low=2006)
    
    if plume == 'grey':
        # plot to main temperature panel
        plume_23rd = ax_main_fig.fill_between(te['years'][te['years'][:]>2017],
                                           plot_ll[te['years'][:]>2017],plot_uu[te['years'][:]>2017],
                                           color='grey',alpha=0.3,linewidth=0)
    elif plume == 'blue':
        # plot to main temperature panel
        plume_23rd = ax_main_fig.fill_between(te['years'][te['years'][:]>2017],
                                           plot_ll[te['years'][:]>2017],plot_uu[te['years'][:]>2017],
                                           color='blue',alpha=0.16,linewidth=0)
    if plume == 'purple':
        # plot to main temperature panel
        plume_23rd = ax_main_fig.fill_between(te['years'][(te['years'][:]>2017)*(te['years'][:]<=2100)],
                                           plot_ll[(te['years'][:]>2017)*(te['years'][:]<=2100)],plot_uu[(te['years'][:]>2017)*(te['years'][:]<=2100)],
                                           color='red',alpha=0.16,linewidth=0, clip_on=False)

if plume == 'all':
    # find the relevent ranges for temperature plume
    plot_mean = te['temp_zero']['RCP8.5'][0][1] 
    plot_uu = te['temp_zero']['RCP8.5'][0][2] 
    plot_ll = te['temp_zero']['RCP8.5'][0][0] 
    plot_ul = te['temp_zero']['RCP8.5'][0][4] 
    plot_lu = te['temp_zero']['RCP8.5'][0][3] 
    
    if baseline_choice == 'nearterm':
        plot_mean = express_as_anom(plot_mean, te['years'], base_high=2015, base_low=2006)
        plot_uu = express_as_anom(plot_uu, te['years'], base_high=2015, base_low=2006)
        plot_ll = express_as_anom(plot_ll, te['years'], base_high=2015, base_low=2006)
        plot_ul = express_as_anom(plot_ul, te['years'], base_high=2015, base_low=2006)
        plot_lu = express_as_anom(plot_lu, te['years'], base_high=2015, base_low=2006)

    # plot to main temperature panel
    plume_23rd = ax_main_fig.fill_between(te['years'][te['years'][:]>2017],
                                       plot_ll[te['years'][:]>2017],plot_uu[te['years'][:]>2017],
                                       color='grey',alpha=0.3,linewidth=0)
    
    # find the relevent ranges for temperature plume
    plot_mean_te2 = te2['temp_zero']['RCP8.5'][0][1] 
    plot_uu_te2 = te2['temp_zero']['RCP8.5'][0][2] 
    plot_ll_te2 = te2['temp_zero']['RCP8.5'][0][0] 
    plot_ul_te2 = te2['temp_zero']['RCP8.5'][0][4] 
    plot_lu_te2 = te2['temp_zero']['RCP8.5'][0][3] 
    
    if baseline_choice == 'nearterm':
        plot_mean_te2 = express_as_anom(plot_mean_te2, te2['years'], base_high=2015, base_low=2006)
        plot_uu_te2 = express_as_anom(plot_uu_te2, te2['years'], base_high=2015, base_low=2006)
        plot_ll_te2 = express_as_anom(plot_ll_te2, te2['years'], base_high=2015, base_low=2006)
        plot_ul_te2 = express_as_anom(plot_ul_te2, te2['years'], base_high=2015, base_low=2006)
        plot_lu_te2 = express_as_anom(plot_lu_te2, te2['years'], base_high=2015, base_low=2006)

    # plot to main temperature panel
    plume_te2_23rd = ax_main_fig.fill_between(te2['years'][te2['years'][:]>2017],
                                       plot_ll_te2[te2['years'][:]>2017],plot_uu_te2[te2['years'][:]>2017],
                                       color='blue',alpha=0.16,linewidth=0)
    
    
    # find the relevent ranges for temperature plume
    plot_mean_pur = te['temp_zero']['RCP8.5'][1][1] 
    plot_uu_pur = te['temp_zero']['RCP8.5'][1][2] 
    plot_ll_pur = te['temp_zero']['RCP8.5'][1][0] 
    plot_ul_pur = te['temp_zero']['RCP8.5'][1][4] 
    plot_lu_pur = te['temp_zero']['RCP8.5'][1][3] 

    if baseline_choice == 'nearterm':
        plot_mean_pur = express_as_anom(plot_mean_pur, te['years'], base_high=2015, base_low=2006)
        plot_uu_pur = express_as_anom(plot_uu_pur, te['years'], base_high=2015, base_low=2006)
        plot_ll_pur = express_as_anom(plot_ll_pur, te['years'], base_high=2015, base_low=2006)
        plot_ul_pur = express_as_anom(plot_ul_pur, te['years'], base_high=2015, base_low=2006)
        plot_lu_pur = express_as_anom(plot_lu_pur, te['years'], base_high=2015, base_low=2006)
    
    # plot to main temperature panel
    plume_pur_23rd = ax_main_fig.fill_between(te['years'][(te['years'][:]>2017)*(te['years'][:]<=2100)],
                                       plot_ll_pur[(te['years'][:]>2017)*(te['years'][:]<=2100)],plot_uu_pur[(te['years'][:]>2017)*(te['years'][:]<=2100)],
                                       color='red',alpha=0.16,linewidth=0, clip_on=False)
    
    # calculate position and size of error bars used to show extent of red plume
    #----------------------------------------
    dy = (plot_lu_pur[te['years'][:]==2100] - plot_ul_pur[te['years'][:]==2100]) / 2.
    y = (plot_lu_pur[te['years'][:]==2100] + plot_ul_pur[te['years'][:]==2100]) / 2.
    epur = ax_main_fig.errorbar(2107, y, yerr=dy, fmt='none', ecolor=(242/255., 155/255., 155/255.), elinewidth=3, capsize=0)
    for b in epur[1]:
        b.set_clip_on(False)
    for b in epur[2]:
        b.set_clip_on(False)

    dy = (plot_uu_pur[te['years'][:]==2100] - plot_ll_pur[te['years'][:]==2100]) / 2.
    y = (plot_uu_pur[te['years'][:]==2100] + plot_ll_pur[te['years'][:]==2100]) / 2.
    epur1 = ax_main_fig.errorbar(2107, y, yerr=dy, fmt='none', ecolor='red', alpha=0.16, elinewidth=1, capsize=3)
    for b in epur1[1]:
        b.set_clip_on(False)
    for b in epur1[2]:
        b.set_clip_on(False)
    #----------------------------------------

    # calculate position and size of error bars used to show extent of blue plume
    #----------------------------------------
    dy = (plot_lu_te2[te['years'][:]==2100] - plot_ul_te2[te['years'][:]==2100]) / 2.
    y = (plot_lu_te2[te['years'][:]==2100] + plot_ul_te2[te['years'][:]==2100]) / 2.
    e_te2 = ax_main_fig.errorbar(2105, y, yerr=dy, fmt='none', ecolor=(142/255., 170/255., 255/255.), elinewidth=3, capsize=0)
    for b in e_te2[1]:
        b.set_clip_on(False)
    for b in e_te2[2]:
        b.set_clip_on(False)

    dy = (plot_uu_te2[te['years'][:]==2100] - plot_ll_te2[te['years'][:]==2100]) / 2.
    y = (plot_uu_te2[te['years'][:]==2100] + plot_ll_te2[te['years'][:]==2100]) / 2.
    e_te21 = ax_main_fig.errorbar(2105, y, yerr=dy, fmt='none', ecolor='blue', alpha=0.16, elinewidth=1, capsize=3)
    for b in e_te21[1]:
        b.set_clip_on(False)
    for b in e_te21[2]:
        b.set_clip_on(False)
    #----------------------------------------

    # calculate position and size of error bars used to show extent of grey plume
    #----------------------------------------
    dy = (plot_lu[te['years'][:]==2100] - plot_ul[te['years'][:]==2100]) / 2.
    y = (plot_lu[te['years'][:]==2100] + plot_ul[te['years'][:]==2100]) / 2.
    e1 = ax_main_fig.errorbar(2103, y, yerr=dy, fmt='none', ecolor='grey', alpha=0.6, elinewidth=3, capsize=0)
    for b in e1[1]:
        b.set_clip_on(False)
    for b in e1[2]:
        b.set_clip_on(False)

    dy = (plot_uu[te['years'][:]==2100] - plot_ll[te['years'][:]==2100]) / 2.
    y = (plot_uu[te['years'][:]==2100] + plot_ll[te['years'][:]==2100]) / 2.
    e11 = ax_main_fig.errorbar(2103, y, yerr=dy, fmt='none', ecolor='grey', alpha=0.4, elinewidth=1, capsize=3)
    for b in e11[1]:
        b.set_clip_on(False)
    for b in e11[2]:
        b.set_clip_on(False)
    #----------------------------------------

# create custom legends for main panel
blue_patch = Patch(facecolor='blue', edgecolor=None, alpha=0.16, label='Faster reductions in CO$_2$ emissions')
red_patch = Patch(facecolor='red', edgecolor=None, alpha=0.16, label='No reductions in non-CO$_2$ climate forcing')
grey_patch = Patch(facecolor='grey', edgecolor=None, alpha=0.4, label='Stylized $1.5^\circ$C pathway')
orange_patch = Patch(facecolor=cols['vermillion'], edgecolor=None, alpha=0.4, label='Anthropogenic warming and $\it{likely}$ range')
grey_patch_2 = Patch(facecolor='grey', edgecolor=None, alpha=0.7, label='Observed monthly global mean surface temperature')

# draw legends to plot
if plume == 'all':
    leg_temp_obs = ax_main_fig.legend(handles=[grey_patch, blue_patch, red_patch], loc=(0.633,0.07),fontsize=10)
    ax_main_fig.legend(handles=[grey_patch_2, orange_patch], loc=(0.02,0.85), fontsize=10)
    ax_main_fig.add_artist(leg_temp_obs)
elif plume == 'grey':
    leg_temp_obs = ax_main_fig.legend(handles=[grey_patch], loc=(0.633,0.07),fontsize=10)
    ax_main_fig.legend(handles=[grey_patch_2, orange_patch], loc=(0.02,0.85), fontsize=10)
    ax_main_fig.add_artist(leg_temp_obs)
elif plume == 'purple':
    leg_temp_obs = ax_main_fig.legend(handles=[red_patch], loc=(0.633,0.07),fontsize=10)
    ax_main_fig.legend(handles=[grey_patch_2, orange_patch], loc=(0.02,0.85), fontsize=10)
    ax_main_fig.add_artist(leg_temp_obs)
elif plume == 'blue':
    leg_temp_obs = ax_main_fig.legend(handles=[blue_patch], loc=(0.633,0.07),fontsize=10)
    ax_main_fig.legend(handles=[grey_patch_2, orange_patch], loc=(0.02,0.85), fontsize=10)
    ax_main_fig.add_artist(leg_temp_obs)
    
plt.show()
```
